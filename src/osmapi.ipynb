```python
place_name = "Lansing, Michigan, USA"
graph = osm.graph_from_place(place_name, network_type='all')
```

```python
osm.plot_graph(osm.project_graph(graph))
```

```python

# Get the hospitals in Lansing
tags = {'amenity': 'hospital'}
parks = osm.features_from_place(place_name, tags)
parks.plot()
```
